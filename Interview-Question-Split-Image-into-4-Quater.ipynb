```python
def splitimage(imgPath, getimageQuater=None):
    img = cv2.imread(imgPath, 0)
    df = pd.DataFrame(img)
    rows, cols = df.shape[0], df.shape[1]
    
    r1 = rows//4
    r2 = r1+r1
    r3 = r2+r1
    r4 = r3+r1

    c1 = cols//4

    q1 = df.iloc[:r1,:c1]
    q2 = df.iloc[r1:r2,c1:]

    q3 = df.iloc[r2:r3,:c1]
    q4 = df.iloc[r3:r4,c1:]
    
    
    if getimageQuater == 'q1':
        plt.imshow(q1,)
        plt.title(f"Image - {getimageQuater}")
        plt.show()
        return q1
    elif getimageQuater == 'q2':
        plt.imshow(q2,)
        plt.title(f"Image - {getimageQuater}")
        plt.show()
        return q2
    elif getimageQuater == 'q3':
        plt.imshow(q3,)
        plt.title(f"Image - {getimageQuater}")
        plt.show()
        return q3
    elif getimageQuater == 'q4':
        plt.imshow(q4,)
        plt.title(f"Image - {getimageQuater}")
        plt.show()
        return q4
    elif getimageQuater == 'o':
        plt.imshow(img,)
        plt.title("Original Image")
        plt.show()
        return img
        
```

```python
image_path = 'prediction.jpg'
q1_pixels = splitimage(image_path, getimageQuater='q1')
q2_pixels = splitimage(image_path, getimageQuater='q2')
q3_pixels = splitimage(image_path, getimageQuater='q3')
q4_pixels = splitimage(image_path, getimageQuater='q4')
full_img_pixels = splitimage(image_path, getimageQuater='o')
```
